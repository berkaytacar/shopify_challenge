We can see that the minimum order amount was 90 dollars,  maximum order amount was 704,000 dollars, and therefore, the mean order amount in this data set is $3145.

We can analyze the data further by looking at the order values between different shops. In order to do that, we group the orders based on their `shop_id`'s and we calculate their mean and median order amounts and total items they sold. 

```r
shop_analysis <- sneaker_df %>%
                group_by(shop_id) %>%
                summarize(num_of_orders = n(), 
                          mean_order_amount = mean(order_amount, na.rm = TRUE),
                          mean_total_items = mean(total_items, na.rm = TRUE),
                          median_order_amount = median(order_amount, na.rm = TRUE),
                          median_total_items = median(total_items, na.rm = TRUE),
                         )
shop_analysis
```

```r
summary(shop_analysis$num_of_orders)
```

By looking at the table above, we can see that most sneaker shops received around the same number of orders ranging between 35-68 with a mean and median of 50 orders per shop. However, sneaker shop with `shop_id` 78 has a lot higher average order amount (49213.04 dollars) compared to the order shops. So, we arrange the table to see if there are other shops which have higher `mean_order_amount`.   

```r
shop_analysis %>%
            arrange(desc(mean_order_amount)) %>%
            head()
```

We can see that shops 42 and 78 have a lot higher `mean_order_amount` than the others, and if we look more carefully, we can see that the mean of the total items sold given in `mean_total_items` column is roughly 668 items for shop 42.

```r
df_42 <- sneaker_df %>%
        filter(shop_id == 42) %>%
        arrange(desc(order_amount))
df_42
```

For shop 42, there are 17 credit card orders of 2000 sneakers bought for 704,000 dollars in March 2017 with the exact same customer or user id. This means that shop 42 might have some sort of a deal with user 607 to purchase 2000 items for $704,000.  

```r
df_78 <- sneaker_df %>%
        filter(shop_id == 78) %>%
        arrange(desc(order_amount))
df_78
```

On the other hand, shop 78 had orders worth 154,350 and 102,900 dollars, but the last observations above tell us that shop 78 sold 1 item for 25,725 dollars to different customers. This suggests that shop 78 could be selling more luxury or higher end products than the other shops. 

These expensive orders placed on shops 42 and 78 cause our analysis of average order value (AOV) to be skewed, and result in AOV of $3145.15. 

### Alternate Option

Because of the fact that there are orders purchased in bulk or maybe higher end or rare sneakers sold by some shops, we get the average order value (AOV) to be skewed by these transactions. Therefore, a good way for us to evaluate this data is to use the median or the mean of the order value per item. 

In order to do that, we need to create a new column called `order_val_per_item` and for each order id, we need to divide the `order_amount` by `total_items`.

```r
mutated_df <- sneaker_df %>%
            mutate(order_val_per_item = order_amount/total_items)
mutated_df %>% head() # look at the head 
dim(mutated_df) # should be 5000x8
```

We can analyze this further by using the summary function after we select the columns we need.  

```r
df <- mutated_df %>% select(order_id, shop_id, user_id, order_val_per_item)
dim(df) # should be 5000x4
head(df)
```

```r
summary(df$order_val_per_item)
```

We can see that the median of the order value per sneaker sold is 153 dollars, and the mean of the order value per sneaker sold is 387.7 dollars. This shows that the shop_id= 78 still affects our mean value quite a bit since it only sells one model of shoe from 25,725 dollars shown below.

```r
mutated_df_analysis <- mutated_df %>%
                group_by(shop_id) %>%
                summarize(num_of_orders = n(), 
                          mean_order_val_per_item = mean(order_val_per_item, na.rm = TRUE)
                         )
shop_78_analysis <- filter(mutated_df_analysis, shop_id == 78)
shop_78_analysis
```

In this case, we could remove the orders with `shop_id` = 78, and note this shop id to further investigate to see if this shop sells luxury sneakers or some other rare items. After removing the outlier, we can check the average order value per item. 

```r
df_78_removed <- df %>%
        filter(shop_id != 78)
summary(df_78_removed$order_val_per_item)
```

When we remove the observations or orders sold by shop id 78, we get the mean of our order value per item as 152.5 dollars which is almost the same as the median of our order value per item (153 dollars).

### Conclusion

Therefore, we can conclude that the best way to evaluate this data to is to look at the `median of average order value per item`. This gives us a lot better idea about the prices of sneakers sold on Shopify, because median isn't much affected by the extreme cases within our data such as the 17 orders of 2000 sneakers or the sneakers sold by `shop_id` 78.

The value for this `median of average order value per item` metric used in this data set is $153.
